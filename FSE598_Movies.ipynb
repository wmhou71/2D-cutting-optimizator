```python
def Sort(Orignal_List, sort_item):
    #sort_item = 'Year'
    sort_mapping = {}
    sort_list = []
    Orignal_List = Movies
    New_List = []

    for index, movie in enumerate(Orignal_List):
        for movie_info in movie['Movie']:
            try:
                sort_list.append(movie_info[sort_item])
                sort_mapping[movie_info[sort_item]] = index
                #print(index, movie_info)
            except KeyError:
                continue
    sort_list = sorted(sort_list)
    #print(sort_list)
    #print(sort_mapping)

    for i in range(len(sort_list)):
        New_List.append(Orignal_List[sort_mapping[sort_list[i]]])
    #print("NEW:", New_List)
    j = json.dumps(New_List, indent = 4)
    #print(j)
    return j
```
